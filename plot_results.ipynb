```python
%matplotlib notebook
import numpy as np
import matplotlib
matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
correct_var_bay1 = np.load("bayesian1_correct_var.npy").mean(1)
incorrect_var_bay1 = np.load("bayesian1_incorrect_var.npy").mean(1)
correct_var_bay2 = np.load("bayesian2_correct_var.npy").mean(1)
incorrect_var_bay2 = np.load("bayesian2_incorrect_var.npy").mean(1)
correct_var_prob = np.load("probabilistic_correct_var.npy").mean(1)
incorrect_var_prob = np.load("probabilistic_incorrect_var.npy").mean(1)
```

```python
print(correct_var_bay1.mean())
print(incorrect_var_bay1.mean())
print(incorrect_var_bay1.mean()/correct_var_bay1.mean())
print()
print(correct_var_bay2.mean())
print(incorrect_var_bay2.mean())
print(incorrect_var_bay2.mean()/correct_var_bay2.mean())
print()
print(correct_var_prob.mean())
print(incorrect_var_prob.mean())
print(incorrect_var_prob.mean()/correct_var_prob.mean())
```

```python
fig, axs = plt.subplots(1, 3, figsize=(6, 2.5), sharey=False)

axs[0].hist([correct_var_bay1, incorrect_var_bay1], bins=20, density=True, alpha=1.0, label=['true', 'false'])
axs[0].semilogy()
#axs[0].set_title(r'bayesian1')
axs[0].set_xlabel(r'uncertainty $ \sigma^2 $')
axs[0].set_ylabel(r'rel. frequency')
axs[0].legend()

axs[1].hist([correct_var_bay2, incorrect_var_bay2], bins=20, density=True, alpha=1.0, label=['true', 'false'])
axs[1].semilogy()
#axs[1].set_title(r'bayesian2')
axs[1].set_xlabel(r'uncertainty $ \sigma^2 $')
#axs[1].set_ylabel(r'prop. density $ P(I) $')
axs[1].legend()

axs[2].hist([correct_var_prob, incorrect_var_prob], bins=20, density=True, alpha=1.0, label=['true', 'false'])
axs[2].semilogy()
#axs[2].set_title(r'probabilistic')
axs[2].set_xlabel(r'uncertainty $ \sigma^2 $')
#axs[2].set_ylabel(r'prop. density $ P(I) $')
axs[2].legend()

plt.tight_layout()
plt.show()
plt.savefig("histograms.pdf", dpi=300)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(6, 2))

axs[0].boxplot([correct_var_bay1, incorrect_var_bay1], sym="+")
axs[0].set_title(r"bayesian1")
axs[0].set_ylabel(r"uncertainty $ \sigma^2 $")
axs[0].set_xticklabels([r'true', r'false'])
ylim = axs[0].get_ylim()
axs[0].set_ylim([ylim[0], 0.0065])

axs[1].boxplot([correct_var_bay2, incorrect_var_bay2], sym="+")
axs[1].set_title(r"bayesian2")
#axs[1].set_ylabel("uncert")
axs[1].set_xticklabels([r'true', r'false'])

axs[2].boxplot([correct_var_prob, incorrect_var_prob], sym="+")
axs[2].set_title(r"variational")
#axs[2].set_ylabel("uncert")
axs[2].set_xticklabels([r'true', r'false'])
ylim = axs[0].get_ylim()
axs[2].set_ylim([ylim[0], 0.0082])

plt.tight_layout()
plt.show()
plt.savefig("boxplots.pdf", dpi=300)
```
